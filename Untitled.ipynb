```python
from pydataset import data
import pandas as pd
from scipy import stats
```

```python
cake = data('cake')
cake[cake.recipe == 'C']
```

```python
atemp = cake.temperature[cake.recipe == 'A']
btemp = cake.temperature[cake.recipe == 'B']
ctemp = cake.temperature[cake.recipe == 'C']
atemp.var(), btemp.var(), ctemp.var()
```

```python
f, p = stats.f_oneway(atemp, btemp, ctemp)
f, p
```
